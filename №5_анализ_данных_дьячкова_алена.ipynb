###Выведите количество выживших мужчин и женщин на корабле - 1 балл
hint: используйте методы: groupby(), agg(), pl.col(), sum()...


```python
survived_by_sex = (
    df.group_by("Sex")
    .agg(pl.col("Survived").sum().alias("Выжившие"))
)
print("Количество выживших мужчин и женщин")
print(survived_by_sex)
```

###Другой способ определения эффективности оптимизации (по зптраченной памяти, а не времени выполнения)

```python
optimized_dtypes = {
    "price": "int32",
    "area": "int16",
    "bedrooms": "int8",
    "bathrooms": "int8",
    "stories": "int8",
    "mainroad": "category",
    "guestroom": "category",
    "basement": "category",
    "hotwaterheating": "category",
    "airconditioning": "category",
    "parking": "int8",
    "prefarea": "category",
    "furnishingstatus": "category"
}

memory_before = df.memory_usage(deep=True).sum()
print(f"Потребление памяти до оптимизации: {memory_before / 1024:.2f} KB")

df_optimized = df.astype(optimized_dtypes)

memory_after = df_optimized.memory_usage(deep=True).sum()
print(f"Потребление памяти после оптимизации: {memory_after / 1024:.2f} KB")

print(f"Сэкономлено памяти: {(memory_before - memory_after) / 1024:.2f} KB")
```
