##Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame - 2 балла
hint: используйте метод requests.get(), используйте в ссылке запроса параметр hasTitles для фильтрации

```python
houses_list = requests.get(
    f'{api_base_url}/houses',
    headers={'Authorization': f'Bearer {auth_token}'}
).json()

df_houses = pd.DataFrame(houses_list)
df_houses.head()
```

```python
# URL API для домов с фильтром по девизу
url_houses_with_motto = "https://www.anapioficeandfire.com/api/houses/"

houses_with_motto = []
page = 1

for i in range (1, len(houses_list)+1):
    response = requests.get(url_houses_with_motto, params={ "hasTitles": "true"})
    houses = response.json()
    houses_with_motto.extend(houses)
    page += 1

# Преобразуем в DataFrame
df_houses_with_motto = pd.DataFrame(houses_with_motto)
df_houses_with_motto.head()
```
